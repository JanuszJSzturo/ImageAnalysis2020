# Basic
Basic usage of OpenCV, Matplotlib and Numpy

```python
import cv2 # OpenCV
import matplotlib.pyplot as plt
import numpy as np
```

### Read an image from your computer

Image is called *building.png* and it is located in a directory called *input*

```python
img = cv2.imread("input/building.png") # Img is loaded as a numpy matrix

# We can describe the matrix
print("Shape")
print(img.shape)

print("Rows")
print(img.shape[0])

print("Columns")
print(img.shape[1])


print("Dimensions")
print(img.shape[2])
```

### Display the image using matplotlib

```python
plt.imshow(img);
```

```python
# We can also select one of the bands and display it
# Each band contain a dimension

band_0 = img[:,:, 0] # you can change between band 0,1 and 2

plt.imshow(band_0, cmap="gray");
```

### Apply operations on the image
We can change the space color, see this [link](https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_colorspaces/py_colorspaces.html) for more information

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

hue = hsv[:,:,0] #hue channel selection of the hsv colorspace
plt.imshow(hsv[:,:,0], cmap="gray");
```

**Matplotlib interpets the RGB images as BGR**

In order to see the original image properly we need to apply a color transform


```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img);
```

## Select regions using numpy

We can clip the image by selecting the regions of interest(known as roi). For example if we want to remove the sky region we can select rows from 350 to 885 and all the image columns.

When we select an image region we don't copy it

```python
roi = img[350:885,:,:] #We selected rows from 350 to 885, all columns and all dimensions 
plt.imshow(roi);
```

**For example we can "paint" the sky area with another color.**

```python
sky_roi = img[0:350,:,:]
sky_roi[:] = (255,91,165) # same code sky_roi[:,:,:]
plt.imshow(img);
```

### Save an image to your computer

You can save the results by using the *imwrite* function of Opencv.

If you construct more complex plots using Matplotlib you can use the *savefig* function.

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

cv2.imwrite("result.png",img);
```
